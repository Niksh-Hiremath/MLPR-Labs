## 1. Code


### Importing libraries


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### Generating synthetic data


```python
SIZE = 100  # Size of each cluster
np.random.seed(0)  # Used for reproducibility of results
```

```python
# mean (mu) and covariance (sigma) matrices for each cluster
mu1 = [2, 2]
sigma1 = [[0.9, -0.0255], [-0.0255, 0.9]]

mu2 = [5, 5]
sigma2 = [[0.5, 0], [0, 0.3]]

mu3 = [-2, -2]
sigma3 = [[1, 0], [0, 0.9]]

mu4 = [-4, 8]
sigma4 = [[0.8, 0], [0, 0.6]]
```

```python
# Generating synthetic data by drawing samples from each distribution
samples1 = np.random.multivariate_normal(mu1, sigma1, SIZE)
samples2 = np.random.multivariate_normal(mu2, sigma2, SIZE)
samples3 = np.random.multivariate_normal(mu3, sigma3, SIZE)
samples4 = np.random.multivariate_normal(mu4, sigma4, SIZE)
```

```python
# Stacking data points drawn from the four gaussian distributions
data = np.vstack([samples1, samples2, samples3, samples4])
```

### Visualizing the data


```python
# Plotting the generated data
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Data")
plt.grid()
plt.show()
```

### Initiating Within-Cluster Sum of Squares (WCSS)


```python
wcss = []  # Initiating Within-Cluster Sum of Squares (WCSS)
```

### Calculating WCSS for different values of $K$


```python
# Calculating WCSS for different values of K from 1 to 20
for k in range(1, 21):
    kmeans = KMeans(n_clusters=k, random_state=0)  # Initializing KMeans with k clusters
    kmeans.fit(data)  # Fitting the model with data
    wcss.append(kmeans.inertia_)  # Storing the WCSS score
    print(f"K = {k:>2} | WCSS = {wcss[-1]:>8.2f}")
```

### Elbow Method Plot


```python
# Plotting the Elbow Method graph
plt.plot(range(1, 21), wcss, marker="o")
plt.xlabel("Number of clusters ($K$)")
plt.ylabel("WCSS Score")
plt.title("Elbow Method to find Optimal $K$")
plt.xticks(range(1, 21))
plt.grid()
plt.show()
```

From the above elbow plot, we can see that the optiminal number of clusters is $K = 4$.


### Performing K-Means Clustering with $K = 4$


```python
# Initializing KMeans with K = 4 and fitting the model with data
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(data)
```

```python
# Predicting cluster labels for each data point
labels = kmeans.predict(data)
labels
```

### Visualizing the clusters


```python
# Visualizing the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.title("Clustered Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid()
plt.show()
```

### Q1: What might happen if the value of the number of clusters is set too high or too low in K-Means?

**Ans:** If the value of $K$ is set too high, K-Means will overfit on the data, meaning it will create too many smaller clusters that may not represent anything meaningful. This will lead to specilization of clusters, where each cluster contains only a few data points. It will not generalize well to new data. It is also highly prone to outliers.

If the value of $K$ is set too low, K-Means will underfit on the data, meaning it will create too few clusters that may combine different groups into a single cluster. This will lead to loss of important information. The clusters will be too broad and not capture the underlying structure of the data.

---


### Q2: What is the effect of initializing centroids far apart or too close together in K-Means, and how might this affect the final clustering?

**Ans:** If the centroids are initialized too close together, then 2 or more centroids may end up in the same cluster, leading to poor clustering results. This will lead the algorithm to create a few smaller clusters close to each other, while creating larger clusters elsewhere to capture the rest of the data points. It causes the model to not capture the underlying structure of the data effectively.

If the centroids are initialized far apart, then they are more likely to be in different clusters, leading to better clustering results. This will help the algorithm to create well-separated clusters that capture the underlying structure of the data effectively.

### Q3: How might you optimise the centroid initialisation to lead to better clustering?

**Ans:** We can use the K-Means++ algorithms to optimize the centroid initialization. This method selects initial centroids that are far apart from each other. This leads to a pseudo-random start. It works by:

- It places the first centroid on top of a randomly selected datapoint in a corner.

- Then, it places the second centroid on a datapoint that is as far away as possible from the first centroid.

- Similarly, it places the third centroid on a datapoint that is as far away as possible from the first two centroids... and so on.

This helps in putting initial centroids around all the corners formed by the datapoints.

### Q4: Why might the Elbow Method not always provide a clear solution for choosing the optimal number of clusters?

**Ans:** We pick the number of clusters where the graph tends to flatten out and form an elbow. However, sometimes the elbow point is not apparent because the decrease in WCSS is gradual and does not show a clear elbow. This usually happens when clusters overlap or the data has noise.

---


### Q5: How can WCSS be influenced by the presence of outliers or noise in the data?

**Ans:** WCSS depends on the square of the distance between data points and their respective centroids. Outliers or noise can significantly increase these distances, leading to a higher WCSS value. This causes the centroids to be pulled towards the outliers, which can lead to poor clustering results.
